# GitHub for Complete Beginners
### A step-by-step guide for data science students

---

## What is GitHub and why should you care?

Imagine you are writing a long essay. You save it as:

```
essay_final.docx
essay_final_v2.docx
essay_final_ACTUALLY_FINAL.docx
essay_final_ACTUALLY_FINAL_fixed.docx
```

Everyone has done this. It is a mess.

**Git** solves this. It tracks every change you make to your files automatically,
so you always have the full history and can go back to any version.

**GitHub** is the website where you store that history online — like Google Drive,
but built specifically for code, with powerful tools for sharing and collaboration.

---

### Two words you will hear constantly

| Word | What it means | Real-world analogy |
|------|--------------|--------------------|
| **Git** | The tool that tracks changes on your computer | Like "track changes" in Word, but for code |
| **GitHub** | The website that stores your Git history online | Like Google Drive, but for code projects |

> Git and GitHub are different things. Git works on your computer. GitHub is a website.
> You use Git to talk to GitHub.

---

### Key vocabulary — learn these 6 words

| Word | Meaning |
|------|---------|
| **Repository (repo)** | A folder that Git is tracking — contains all your project files and their full history |
| **Commit** | A saved snapshot of your files at a point in time — like a checkpoint in a game |
| **Clone** | Download a copy of a repo from GitHub to your computer |
| **Fork** | Make your own personal copy of someone else's repo on GitHub |
| **Push** | Upload your local commits to GitHub |
| **Pull** | Download the latest changes from GitHub to your computer |

```
The flow in one picture:

  GitHub (online)          Your Computer (local)
  ───────────────          ─────────────────────
  
  [Original repo]  →fork→  [Your GitHub copy]
                                      │
                                   clone↓
                              [Files on your laptop]
                                      │
                              edit → commit
                                      │
                                   push↑
                              [Your GitHub copy updated]
```

## Step 1 — Create a GitHub account

1. Go to **https://github.com**
2. Click **Sign up** in the top right corner
3. Enter your email address
4. Create a password
5. Choose a username
   > ⚠️ **Pick a professional username** — this will be visible to future employers.
   > Use something like `firstname-lastname` or `firstnamelastname`.
   > Avoid usernames like `coolguy99` or `xXdarkcoderXx`.
6. Complete the verification puzzle
7. Check your email and click the confirmation link

That is it — your account is ready.

---

Once logged in, go to **github.com/your-username** — this is your public profile.  
Employers and collaborators will look at this page. It shows:
- All your public repositories
- Your activity (how often you commit code)
- Projects you have contributed to

> 💡 Think of your GitHub profile as your coding CV. Keep it tidy.

## Step 2 — Install Git on your computer

GitHub is the website. Git is the tool you install locally to talk to it.

### Check if Git is already installed

Open a terminal (Mac/Linux) or Command Prompt (Windows) and type:

```bash
git --version
```

If you see something like `git version 2.39.0` → Git is already installed, skip to Step 2a (VS Code login).  
If you see `command not found` → follow the install steps below, then continue to Step 2a.

---

### Install Git

**Windows:**
1. Go to https://git-scm.com/download/win
2. Download and run the installer
3. Click Next through all the steps (defaults are fine)
4. When done, open **Git Bash** from the Start menu — use this instead of Command Prompt

**Mac:**
1. Open Terminal
2. Type: `xcode-select --install`
3. Click Install in the popup — this installs Git automatically

**Linux (Ubuntu/Debian):**
```bash
sudo apt update
sudo apt install git
```

## Step 2a — Connect VS Code to GitHub

Now that Git is installed, connect VS Code to your GitHub account.
This lets you push and pull using clicks instead of terminal commands — and you only do it once.

### Login through VS Code

1. Open VS Code
2. Click the **Accounts icon** at the bottom left of the sidebar (looks like a person)
3. Click **Sign in with GitHub**
4. A browser window opens → log into GitHub → click **Authorize Visual Studio Code**
5. Switch back to VS Code — you are now connected

```
Bottom left of VS Code:
  ┌─────────────────────┐
  │  ⚙  Manage         │
  │  👤 Sign in  ← here │
  └─────────────────────┘
```

---

### The Source Control panel — Git with clicks

Once logged in, press **Ctrl+Shift+G** (or click the branch icon on the left sidebar)
to open the Source Control panel. This is a visual interface for Git — no terminal needed.

```
Source Control panel:
  ┌──────────────────────────────────┐
  │  Message: 'Completed exercise 3' │  ← type commit message here
  │  [ ✓ Commit ]                    │  ← click to commit
  │                                  │
  │  Changes                         │
  │    M  week3_notebook.ipynb       │  ← M = modified file
  │    U  new_file.py                │  ← U = new untracked file
  └──────────────────────────────────┘
```

| What you want to do | In the panel |
|---------------------|-------------|
| Stage files | Click **+** next to each file, or **+** next to Changes for all |
| Commit | Type a message in the box → click **✓ Commit** |
| Push | Click **··· → Push** or the sync icon at the bottom status bar |
| Pull | Click **··· → Pull** |

> 💡 The Source Control panel and the terminal commands do exactly the same thing.
> The panel is easier to start with — the terminal gives you more control later.
> We will learn both.

## Step 2b — Tell Git who you are (terminal setup)

Whether you use the VS Code panel or the terminal, Git needs to know your name and email.
Open the terminal in VS Code (**Ctrl + `**) and run these two commands once:
```bash
git config --global user.name "Your Name"
git config --global user.email "your@email.com"
```

Replace with your own name and the same email you used for GitHub.  
You only do this once — Git remembers it permanently and attaches it to every commit you make.

## Step 3 — Fork the class repository

Someone has a repository you want to work from — a team project, a course, an open source project, or your own existing repo.  
**Forking** creates your own personal copy of it on GitHub — you can edit your copy
without affecting the original.

### What is the difference between Fork and Clone?

| | Fork | Clone |
|--|------|-------|
| **Where** | On GitHub (online) | On your computer (local) |
| **What it creates** | Your own GitHub copy of the repo | A local folder with the files |
| **When to use** | When you want your own version of someone else's project | When you want to work on a repo locally |
| **Typical flow** | Fork first → then Clone your fork | |

---

### How to Fork

1. Go to the repository URL you want to fork
2. In the top-right corner, click the **Fork** button

```
  ┌─────────────────────────────────────────┐
  │  owner-username / repo-name    │
  │                          [Fork]  ← click │
  └─────────────────────────────────────────┘
```

3. GitHub asks "Where should we fork this to?" — choose your own account
4. Click **Create fork**
5. After a few seconds you will see the repo appear under YOUR username:
   `github.com/your-username/repo-name`

> ✅ You now have your own copy. Changes you make here will NOT affect the original repo's.

---
## Step 4 — Clone the repo to your computer

Forking created a copy on GitHub. Now you need to download it to your laptop.

### Get the repo URL

1. Go to YOUR forked repo on GitHub (`github.com/your-username/repo-name`)
2. Click the green **< > Code** button
3. Make sure **HTTPS** is selected
4. Click the copy icon to copy the URL — it looks like:
   `https://github.com/your-username/repo-name.git`

---

### Run the clone command

Open your terminal, navigate to where you want the folder to live, then run:

```bash
# First, go to the folder where you want to put the project
# Example: go to your Desktop
cd ~/Desktop

# Then clone — paste YOUR repo URL here
git clone https://github.com/your-username/repo-name.git
```

You will see output like:
```
Cloning into 'repo-name'...
remote: Enumerating objects: 24, done.
remote: Counting objects: 100%, done.
Receiving objects: 100%, done.
```

A new folder called `repo-name` now exists on your computer with all the files.

```bash
# Enter the folder
cd repo-name

# List the files to confirm everything is there
ls
```

> 💡 **cd** = "change directory" — it moves you into a folder in the terminal.  
> **ls** = "list" — it shows what is inside the current folder.

## 🆕 Starting a brand new project

The steps above (Fork → Clone) are for working on an **existing** repo.
If you are starting from scratch, use one of these two options instead.

---

### Option A — repo already exists on GitHub

The easiest way. Download it — this creates the folder and links it to GitHub automatically:

```bash
git clone https://github.com/your-username/REPO_NAME.git
cd REPO_NAME
```

Done. You are ready to work.

---

### Option B — you already have a local folder and want to link it to a new GitHub repo

Use this when you started working locally first and now want to put it on GitHub.

**First:** go to github.com → click **+** (top right) → **New repository** → create it **empty** (no README, no .gitignore).

**Then** run these commands in VS Code terminal:

```bash
# 1. Go into your existing folder
cd C:\path\to\your\folder

# 2. Turn this folder into a Git repository
git init

# 3. Link it to the empty repo you just created on GitHub
git remote add origin https://github.com/your-username/REPO_NAME.git

# 4. Stage all your existing files
git add .

# 5. Make the first commit
git commit -m "First commit"

# 6. Rename the branch to main (GitHub's default)
git branch -M main

# 7. Push and link local main to GitHub main (-u only needed the first time)
git push -u origin main
```

After this, future pushes are just `git push` — no need for `-u origin main` again.

---

### Which option should I use?

| Situation | Use |
|-----------|-----|
| Repo already exists on GitHub (someone else's, or you created it there first) | Option A — `git clone` |
| You have files on your computer and want to upload them to a new repo | Option B — `git init` |

## Step 6 — Save your work to GitHub (commit + push)

This is the most important habit to build. After finishing work, run these 3 commands.
You will use them every single time.

```bash
# 1. Check what files you changed
git status
```

You will see something like:
```
On branch main
Changes not staged for commit:
  modified:   week3_distributions.ipynb
```

```bash
# 2. Stage the files you want to save
#    The dot (.) means "add ALL changed files"
git add .

# 3. Commit — save a snapshot with a message describing what you did
git commit -m "Completed week 3 distribution exercises"

# 4. Push — send the commit to GitHub
git push
```

After push, go to `github.com/your-username/repo-name` and refresh —
your changes are now online.

---

### The 3-command habit — memorise this

```bash
git add .
git commit -m "your message here"
git push
```

Run these every time you finish a working session. Think of it as pressing Save.

---

### How to write good commit messages

The message in `git commit -m "..."` is a note to your future self.

|  Bad message | Good message |
|---------------|----------------|
| `"stuff"` | `"Add Poisson exercise solutions"` |
| `"fix"` | `"Fix TypeError in Normal distribution cell"` |
| `"changes"` | `"Complete week 3 hypothesis testing notebook"` |
| `"asdfjkl"` | `"Add comments explaining pmf vs cdf"` |

**Rules for good commit messages:**
- Start with a verb: Add, Fix, Complete, Update, Remove
- Be specific enough that you know what you did 3 months from now
- Keep it under 72 characters

## Step 7 — Get updates from the original repo

When the original repo gets new commits — new files, bug fixes, updates — you need to pull those
changes into your fork. You have two options depending on how comfortable you are with the terminal.

---

### Option A — Sync fork on GitHub (easiest, no terminal needed)

GitHub has a built-in button that does everything for you.

1. Go to YOUR fork on GitHub — `github.com/your-username/repo-name`
2. GitHub will show a notice when you are behind:
```
This branch is 2 commits behind owner-username/repo-name
                                        [Sync fork] ← click this
```
3. Click **Sync fork** → click **Update branch**
4. Then in VS Code terminal run:
```bash
git pull
```

That's it — one button click and one command.

```
Original repo updated
        ↓
[Sync fork] on GitHub     ← click online
        ↓
git pull                  ← run in VS Code terminal
        ↓
New files appear on your computer
```

---

### Option B — Sync via terminal only

Use this if you are comfortable with the terminal or need more control.

**One-time setup — add the original repo as upstream:**

```bash
# Run this ONCE after cloning
git remote add upstream https://github.com/owner-username/repo-name.git

# Verify it worked — you should see both origin (yours) and upstream (original)
git remote -v
```

You should see:
```
origin    https://github.com/your-username/repo-name.git   (fetch)
origin    https://github.com/your-username/repo-name.git   (push)
upstream  https://github.com/owner-username/repo-name.git  (fetch)
upstream  https://github.com/owner-username/repo-name.git  (push)
```

**Every time the original repo is updated:**

```bash
# 1. Fetch the latest changes from the original repo
git fetch upstream

# 2. Merge those changes into your local copy
git merge upstream/main

# 3. Push the updated files to YOUR GitHub fork
git push
```

---

> 💡 **Which should I use?**  
> Starting out → Option A. It is faster and harder to break.  
> When you are more confident → Option B gives you more visibility into what changed.

## Quick Reference — commands you will use every week

```bash
# ── DAILY WORKFLOW ──────────────────────────────────────────────────────────

# 1. Before starting: get latest updates from the original repo
git fetch upstream
git merge upstream/main

# 2. Check what you have changed
git status

# 3. After finishing: save to GitHub
git add .
git commit -m "Describe what you did"
git push


# ── USEFUL EXTRAS ───────────────────────────────────────────────────────────

# See your commit history
git log --oneline

# See exactly what changed in a file
git diff filename.ipynb

# Undo ALL unsaved changes to a file (careful — cannot be undone!)
git checkout -- filename.ipynb

# Go back to how a file looked in a previous commit
git checkout abc1234 -- filename.ipynb   # abc1234 = commit ID from git log
```

## Best practices — habits that will save you

**1. Commit often, not just at the end**  
Commit after every meaningful chunk of work — finished an exercise, fixed a bug, added comments.
Small commits are easier to understand and easier to undo.

**2. Always pull before you push**  
Before starting a session, always fetch and merge from upstream.
This prevents conflicts where two people changed the same file.

**3. Never commit passwords or API keys**  
If a file contains a password or secret key, add it to a `.gitignore` file first.
Once something is on GitHub it is public and permanent — even if you delete it later.

**4. Read git status before git add**  
Always run `git status` first so you know exactly what you are about to commit.
It prevents accidentally committing files you did not mean to.

**5. Write your commit message as if explaining to a colleague**  
"Fixed it" tells nobody anything. "Fix KeyError in Poisson pmf exercise" is useful.

## Common errors and how to fix them

| Error message | What it means | Fix |
|---------------|--------------|-----|
| `fatal: not a git repository` | You are not inside a Git folder | `cd` into your project folder first |
| `nothing to commit, working tree clean` | No changes since last commit | Nothing to do — you are up to date |
| `Please tell me who you are` | Git config not set | Run the `git config` commands from Step 2 |
| `rejected — fetch first` | Someone else pushed changes you don't have | Run `git fetch upstream && git merge upstream/main` first |
| `CONFLICT (content)` | Two people edited the same lines | Open the file, look for `<<<<<<<` markers, fix manually, then commit |
| `Permission denied (publickey)` | SSH not set up | Use HTTPS URLs instead of SSH (starts with `https://`, not `git@`) |
